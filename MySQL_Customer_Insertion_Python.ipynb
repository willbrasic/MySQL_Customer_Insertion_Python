# Database Customer Insertion
## This program's purpose is to insert customer information into a MySQL database. 

Input: 
1. Customer Information

Output: 
1. New Database Record 

Processing:
1. Determine if the database already has the customer the user enters.
2. Update the database if (1.) is false. 

```python
## importing necessary modules ##
import mysql.connector

## create a connect object ##
connect = mysql.connector.connect(
  host="localhost", # server
  user="root", # user name
  passwd="" #password
)

## creating a cursor ##
cursor = connect.cursor()

## using the CoffeeDB database ##
cursor.execute('use CoffeeDB')

## prompt user to enter customer number ##
cust_num = input('Please enter the customer number:  \n').strip()
while (cust_num == ''):
    print('The customer number cannot be empty. Please enter again:  ')
    cust_num = input().strip()

## input validation for length of customer number ##
while (len(cust_num) != 3):
    cust_num = input('The customer number can only be three digits. Please enter a valid entry:  ').strip()

## query to get all customer numbers ##
cursor.execute(
        '''SELECT CustomerNumber 
        FROM customer'''
)

## adding all the customer numbers to a list (figured out this fancy way to do a loop) ##
existing_cust_nums = [i[0] for i in cursor]

## if the customer already exists display that record ##
if (cust_num in existing_cust_nums):
    print(f'''Customer {cust_num} already exists. 
Here is the current information for this customer:''')
    
    ## query to get the requested customer number ##
    query = f'SELECT * FROM customer WHERE CustomerNumber = "{cust_num}"'
    
    ## execute the above query ##
    cursor.execute(query)
    
    ## displaying customer information ##
    print()
    for i in cursor:
        print('Name:     ', i[1])
        print('Address:  ',  i[2], i[3], i[4], i[5])

## if the customer does not already exist then make a new record ##
else:   
    cust_name = input('Please enter customer\'s name:                       ').title()
    cust_address = input('Please enter the customer\'s address:                ').title()
    cust_city = input('Please enter the customer\'s city:                   ').title()
    cust_state = input('Please enter the customer\'s state abbreviation:     ').title().strip()
    while (len(cust_state) != 2):
        cust_state = input('The customer state must be only 2 letters. Please enter again:  ').title().strip()
    cust_zip = input('Please enter the customer\'s zip code:               ').strip()
    
    ## query to insert the data ##
    query = 'INSERT INTO customer VALUES (%s, %s, %s, %s, %s, %s)'

    ## values to be inserted. must match the column order in the database ##
    values = (cust_num, cust_name, cust_address, cust_city, cust_state, cust_zip)

    ## execute the above query with the values provided ##
    cursor.execute(query, values)

    ## show a confirmation message to the user displaying how many rows have been inserted ##
    print()
    print(str(cursor.rowcount) + ' row inserted.')

    ## commit the changes to the database ##
    connect.commit()

    ## close the connection ##
    connect.close ()
```
